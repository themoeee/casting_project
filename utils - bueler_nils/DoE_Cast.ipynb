```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sampleGenerator import LHSampler, SampleGenerator
from itertools import product
```

```python
def round_linear(x, base=.05):
  return (base * (np.array(x) / base).round())

def round_log(x, base=10):
  return (base ** ((np.log10(np.array(x))/np.log10(base)).round()))

def round_list(x, lst):
  arr = np.array(x)
  out = np.zeros(arr.shape)
  for i in range(len(arr)):
    out[i] = min(lst, key=lambda y: abs(y - arr[i]))
  return out
```

First phase parameters

```python
type_1st_phase = {"m": 1.75, "s": 1.25, "dist": "uniform", "round": round_linear, "base": 1}
s1 = {"m": 200, "s": 100, "dist": "uniform", "round": round_linear, "base": 20}
v1 = {"m": 0.3, "s": 0.2, "dist": "uniform", "round": round_linear, "base": 0.1}
s2_s1 = {"m": 200, "s": 100, "dist": "uniform", "round": round_linear, "base": 20}  # s2 - s1
v2_v1 = {"m": 2, "s": 1, "dist": "uniform", "round": round_linear, "base": 1/3}     # v2/v1
vcrit = {"m": 0.65, "s": 0.35, "dist": "uniform", "round": round_linear, "base": 0.1}   # ratio to v_crit
```

Second phase parameters

```python
vI = {"m": 3, "s": 1, "dist": "uniform", "round": round_linear, "base": 0.5}
s3_shift = {"m": 25, "s": 75, "dist": "uniform", "round": round_linear, "base": 20}     # shift to s_ma
v3_vI_v2 = {"m": 0.5, "s": 0.5, "dist": "uniform", "round": round_linear, "base": 0.1}  # ratio of v3/(vI-v2) --> 1 => vI, 0 => v2
s34_shift = {"m": 70, "s": 30, "dist": "uniform", "round": round_linear, "base": 20}    # s4 - s3
sbreak_shift = {"m": -35, "s": 15, "dist": "uniform", "round": round_linear, "base": 10}    # shift to s_brake
vbrake = {"m": 1, "s": 0.5, "dist": "uniform", "round": round_linear, "base": 0.25}
```

Third phase parameters

```python
p0 = {"m": 300, "s": 150, "dist": "uniform", "round": round_linear, "base": 100}
dp1 = {"m": 100, "s": 150, "dist": "uniform", "round": round_linear, "base": 100}
t1 = {"m": 0.065, "s": 0.035, "dist": "uniform", "round": round_linear, "base": 0.01}
dp2 = {"m": 150, "s": 200, "dist": "uniform", "round": round_linear, "base": 100}
dt2 = {"m": 0.06, "s": 0.04, "dist": "uniform", "round": round_linear, "base": 0.02}
sol_time = {"m": 11, "s": 9, "dist": "uniform", "round": round_log, "base": 2}
```

Temperatures

```python
hm2 = {"m": 180, "s": 25, "dist": "normal", "round": round_linear, "base": 20}
hm3 = {"m": 160, "s": 25, "dist": "normal", "round": round_linear, "base": 20}
hm4 = {"m": 160, "s": 25, "dist": "normal", "round": round_linear, "base": 20}
m4 = {"m": 120, "s": 25, "dist": "normal", "round": round_linear, "base": 20}

piston = {"m": 60, "s": 25, "dist": "normal", "round": round_linear, "base": 20}

hf1 = {"m": 160, "s": 25, "dist": "normal", "round": round_linear, "base": 20}
hf2 = {"m": 180, "s": 25, "dist": "normal", "round": round_linear, "base": 20}
hf3 = {"m": 160, "s": 25, "dist": "normal", "round": round_linear, "base": 20}
f3 = {"m": 120, "s": 25, "dist": "normal", "round": round_linear, "base": 20}
f4 = {"m": 120, "s": 25, "dist": "normal", "round": round_linear, "base": 20}

metal = {"m": 690, "s": 20, "dist": "normal", "round": round_linear, "base": 20}
```

Vacuum

```python
pvac = {"m": 6.5, "s": 4, "dist": "uniform", "round": round_list, "base": [10, 20, 50, 100, 200, 500, 1000]}
```

Sampling

```python
params = {
    "type_1st_phase": type_1st_phase,
    "s1": s1,
    "v1": v1,
    "s2_s1": s2_s1,
    "v2_v1": v2_v1,
    "vcrit": vcrit,
    "vI": vI,
    "s3_shift": s3_shift,
    "v3_vI_v2": v3_vI_v2,
    "s34_shift": s34_shift,
    "sbreak_shift": sbreak_shift,
    "vbrake": vbrake,
    "p0": p0,
    "dp1": dp1,
    "t1": t1,
    "dp2": dp2,
    "dt2": dt2,
    "sol_time": sol_time,
    "hm2": hm2,
    "hm3": hm3,
    "hm4": hm4,
    "m4": m4,
    "piston": piston,
    "hf1": hf1,
    "hf2": hf2,
    "hf3": hf3,
    "f3": f3,
    "f4": f4,
    "metal": metal,
    "pvac": pvac
}
```

```python
n_samples = 100
samples = SampleGenerator(LHSampler())(n_samples, [p["m"] for p in params.values()], [p["s"] for p in params.values()], [p["dist"] for p in params.values()])
samples
```

Discretize the values

```python
rounded_samples = np.zeros(samples.shape)
for i, p in enumerate(params.values()):
  rounded_samples[:, i] = p["round"](samples[:, i], p["base"]) if p != pvac else p["round"](2**samples[:, i], p["base"])
rounded_samples
```

```python
samples_df = pd.DataFrame(rounded_samples, columns=params.keys())
samples_df.head()
```

```python
samples_df.describe()
```

Adjust parameters s.t. they can be used directly

```python
samples_df.rename(columns={"s2_s1": "s2", "v2_v1": "v2", "v3_vI_v2": "v3", "dp1": "p1", "dp2": "p2", "dt2": "t2"}, inplace=True)
samples_df["s2"] += samples_df["s1"]
samples_df["v2"] *= samples_df["v1"]
samples_df["v2"] = samples_df["v2"].round(1)
samples_df["v3"] = samples_df["v3"]*(samples_df["vI"] - samples_df["v2"]) + samples_df["v2"]
samples_df["v3"] = samples_df["v3"].round(1)
samples_df["p1"] += samples_df["p0"]
samples_df["p2"] += samples_df["p1"]
samples_df["t2"] += samples_df["t1"]
samples_df.replace({"type_1st_phase": {1: "no optimization", 2: "Buehler any", 3: "Nogowizin"}}, inplace=True)
samples_df
```

Clip temperatures

```python
samples_df["hm2"] = samples_df["hm2"].clip(hm2["m"]-40, hm2["m"]+40)
samples_df["hm3"] = samples_df["hm3"].clip(hm3["m"]-40, hm3["m"]+40)
samples_df["hm4"] = samples_df["hm4"].clip(hm4["m"]-40, hm4["m"]+40)
samples_df["m4"] = samples_df["m4"].clip(m4["m"]-40, m4["m"]+40)

samples_df["piston"] = samples_df["piston"].clip(piston["m"]-40, piston["m"]+40)

samples_df["hf1"] = samples_df["hf1"].clip(hf1["m"]-40, hf1["m"]+40)
samples_df["hf2"] = samples_df["hf2"].clip(hf2["m"]-40, hf2["m"]+40)
samples_df["hf3"] = samples_df["hf3"].clip(hf3["m"]-40, hf3["m"]+40)
samples_df["f3"] = samples_df["f3"].clip(f3["m"]-40, f3["m"]+40)
samples_df["f4"] = samples_df["f4"].clip(f4["m"]-40, f4["m"]+40)

samples_df["metal"] = samples_df["metal"].clip(metal["m"]-30, metal["m"]+30)
```

```python
samples_df.describe()
```

Shot curve generation

```python
# please check - just took the values from the Excel
d_M = 80/1000
V_I = 1527/1e6
V_part = 1025/1e6
V_A = 1120 / 1e6
V_channel = 407/1e6
t_biscuit = 27/1000
l_m_activ = 756/1000
H_biscuit = 30/1000
f_liq_sol = 0.9425

fraction_prefilling = 0
```

```python
A_dM = np.square(d_M)*np.pi/4
FR = V_I/(A_dM*l_m_activ)
V_biscuit = A_dM*t_biscuit

V_prefilling = fraction_prefilling * V_part

s_m_100 = l_m_activ*(1-FR)
s_ma = s_m_100 + (V_channel-V_biscuit)/A_dM
s_pf = s_m_100 + (V_channel-V_biscuit+V_prefilling)/A_dM
s_ov = s_m_100 + (V_channel-V_biscuit+V_part)/A_dM
s_ffin = s_m_100 + (V_channel - V_biscuit + V_A)/A_dM

s_Ifin_solid = s_ffin + V_I * (1-f_liq_sol)/A_dM

v_kr=(9.81*d_M)**0.5*(1.386-1.915*FR+0.616*FR**2) 

# unit to mm
s_m_100 *= 1000
s_ma *= 1000
s_pf *= 1000
s_ov *= 1000
s_ffin *= 1000
s_Ifin_solid *= 1000
```

```python
samples_df.rename(columns={"s3_shift": "s3", "s34_shift": "s4", "sbreak_shift": "s_start_braking"}, inplace=True)
samples_df["s3"] += s_ma
samples_df["s4"] += samples_df["s3"]
samples_df["s_start_braking"] += s_ffin
samples_df["vcrit"] *= v_kr
samples_df = samples_df.round(4)
samples_df.head()
```

Save to csv

```python
samples_df.to_csv("Parameter_combinations_202509.csv")
```
